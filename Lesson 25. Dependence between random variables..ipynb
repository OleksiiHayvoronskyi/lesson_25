### Завдання 1. *Обчислити коефіцієнт кореляції   $ r_{xy} $*.
 
- [x] Ймовірність того, що при перевірці деталь виявиться стандартною, дорівнює $ 0,6 $. Перевірці підлягають $ 3 $ деталі. Побудувати закон розподілу системи двох дискретних випадкових величин $ Х $ – появи числа бракованих деталей і $ Y $ – появи стандартних деталей.
   * Обчислити коефіцієнт кореляції   $ r_{xy} $.

**Наші дані**: $ \\ $ 
$ n = 3 $ (кількість деталей) $ \\ $
$ p = 0,6 $ (ймовірність стандартних деталей) $ \\ $
За формулою: $ q = 1 - p, $ визначаю ймовірність браку $ \\ $
$ q = 1 - 0,6 = 0,4 $ (ймовірність браку)


**Вираховую закон розподілу двох дискретних величин за формулою Бернуллі**: $ \\ $
$ P _m, _n = C^m_n p^m q^{n-m}$, $ \\ $
*де* $ P_0 = C^0_3 ⋅ (0,4)^0 ⋅ (0,6)^3 = 0,216 = \frac{3!}{0! 3!} ⋅ 0,216 = 0,216$ $ \\ $
$ P_1 = C^1_3 ⋅ (0,4)^1 ⋅ (0,6)^2 = 0,144 = \frac{3!}{1! 2!} ⋅ 0,144 = 0,432$ $ \\ $
$ P_2 = C^2_3 ⋅ (0,4)^2 ⋅ (0,6)^1 = 0,096 = \frac{3!}{2! 1!} ⋅ 0,096 = 0,288$ $ \\ $
$ P_3 = C^3_3 ⋅ (0,4)^3 ⋅ (0,6)^0 = 0,064 = \frac{3!}{3! 0!} ⋅ 0,064 = 0,064$ $ \\ $

**Побудую закон розподілу у табличній формі відповідно до обчислених вище даних**: $ \\ $

Вираховую *МАТЕМАТИЧНЕ СПОДІВАННЯ* для $ M(X) $: $ \\ $
$ M(X) = 0 ⋅ 0,216 + 1 ⋅ 0,432 + 2 ⋅ 0,288 + 3 ⋅ 0,064 = 0 + 0,432 + 0,576 + 0,192 = 1,2 $ $ \\ $
$ M(X) = 1,2 $ 

Математичне сподівання КВАДРАТА $ M(X^2) $: $ \\ $
$ M(X^2) = 0^2 ⋅ 0,216 + 1^2 ⋅ 0,432 + 2^2 ⋅ 0,288 + 3^2 ⋅ 0,064 = 0 + 0,432 + 1,152 + 0,576 = 2,16 $ $ \\ $
$ M(X^2) = 2,16 $ 

Визначаю ДИСПЕРСІЮ для $ X $: $ \\ $
$ D(X) = M(X^2) - (M(X)^2) = 2,16 - 1,2^2 = 0,72$ $ \\ $
$ D(X) = 0,72 $ 

СТУПІНЬ РИЗИКУ для $ X $: $ \\ $
$ σ(X) = \sqrt{D(X)} = \sqrt{0,72} ≈ 0,84 $ $ \\ $
$ σ(X) = 0,84 $

Вираховую *МАТЕМАТИЧНЕ СПОДІВАННЯ* для $ M(Y) $: $ \\ $
$ M(Y) = 0 ⋅ 0,064 + 1 ⋅ 0,288 + 2 ⋅ 0,432 + 3 ⋅ 0,216 = 0 + 0,288 + 0,864 + 0,648 = 1,8 $ $ \\ $
$ M(Y) = 1,8 $ 

Математичне сподівання КВАДРАТА $ M(Y^2) $: $ \\ $
$ M(Y^2) = 0^2 ⋅ 0,064 + 1^2 ⋅ 0,288 + 2^2 ⋅ 0,432 + 3^2 ⋅ 0,216 = 0 + 0,288 + 1,728 + 1,944 = 3,96 $ $ \\ $
$ M(Y^2) = 3,96 $ 

Визначаю ДИСПЕРСІЮ для $ Y $: $ \\ $
$ D(Y) = M(Y^2) - (M(Y)^2) = 3,96 - 1,8^2 = 0,72$ $ \\ $
$ D(Y) = 0,72 $ 

СТУПІНЬ РИЗИКУ для $ Y $: $ \\ $
$ σ(Y) = \sqrt{D(X)} = \sqrt{0,72} ≈ 0,84 $ $ \\ $
$ σ(Y) = 0,84 $

Вираховую МАТЕМАТИЧНЕ СПОДІВАННЯ появи ДВОХ ПОДІЙ $ \\ $
$ M(XY) = \sum^n _{i=i} \sum^m _{j=1} y_i x_j p_{ij} $ $ \\ $
$ M(XY) = 2 ⋅ 0,432 + 2 ⋅ 0,288 = 0,864 + 0,576 = 1,44 $ $ \\ $
$ M(XY) = 1,44 $

Визначаю КОВАРІАЦІЮ: $ \\ $
$ cov(X, Y) = M(XY) - M(X) ⋅ M(Y) = 1,44 - 1,2 ⋅ 1,8 = -0,72$ $ \\ $
$ cov(X, Y) = -0,72 $ $ \\ $

Визначаю КОЕФІЦІЄНТ КОРЕЛЯЦІЇ: $ \\ $
$ r(X,Y) = \frac{cov(X, Y)} {σ(X) \mathop ⋅ σ(Y)} = \frac{-0,72} {0,84 \mathop ⋅ 0,84 } ≈ -1 $

**ВІДПОВІДЬ: $ -1 $**
